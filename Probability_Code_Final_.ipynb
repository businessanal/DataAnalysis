```python
import numpy as np
import matplotlib.pyplot as plt

# Set parameters for the exponential distribution
mean_lifetime = 10
num_samples = 1000

# Step 1: Generate the dataset
computer_part_lifetimes = np.random.exponential(scale=mean_lifetime, size=num_samples)

# Step 2: Calculate probability, mean, and variance
probability_more_than_7_years = np.mean(computer_part_lifetimes > 7)
mean_lifetime_sample = np.mean(computer_part_lifetimes)
variance_sample = np.var(computer_part_lifetimes)

# Create a figure with two subplots
fig, axes = plt.subplots(1, 2, figsize=(12, 5))

# Step 3: Create a histogram to visualize the discrete distribution
axes[0].hist(computer_part_lifetimes, bins=30, density=True, alpha=0.6, color='b', label='Lifetime Distribution (Discrete)')
axes[0].axvline(mean_lifetime_sample, color='r', linestyle='dashed', linewidth=2, label='Sample Mean')
axes[0].set_xlabel('Lifetime (years)')
axes[0].set_ylabel('Probability Density')
axes[0].legend()
axes[0].set_title('Exponential Distribution of Computer Part Lifetimes (Discrete)')

# Generate a continuous plot
x = np.linspace(0, max(computer_part_lifetimes), 1000)
y = 1 / mean_lifetime * np.exp(-x / mean_lifetime)

# Step 4: Create a continuous plot to visualize the distribution
axes[1].plot(x, y, color='g', label='Exponential PDF (Continuous)')
axes[1].axvline(mean_lifetime_sample, color='r', linestyle='dashed', linewidth=2, label='Sample Mean')
axes[1].set_xlabel('Lifetime (years)')
axes[1].set_ylabel('Probability Density')
axes[1].legend()
axes[1].set_title('Exponential PDF (Continuous)')

# Display the plots
plt.tight_layout()
plt.show()

# Print the calculated values
print("Problem Statement:")
print("You are analyzing the lifetimes of computer parts. The average lifetime is 10 years, and you want to calculate the probability that a part lasts more than 7 years.")
print("\nResults:")
print(f"Probability that a part lasts more than 7 years: {probability_more_than_7_years:.2f}")
print(f"Sample Mean: {mean_lifetime_sample:.2f}")
print(f"Sample Variance: {variance_sample:.2f}")
```

Uniform Distribution

```python
import numpy as np
import matplotlib.pyplot as plt

# Step 1: Create the dataset
num_participants = 100
tokens = np.arange(1, num_participants + 1)

# Step 2: Calculate Probabilities
probability_less_than_10 = np.mean(tokens <= 10)
probability_even_number = np.mean(tokens % 2 == 0)
probability_between_30_and_70 = np.mean((tokens >= 30) & (tokens <= 70))

# Step 3: Calculate Mean and Variance
mean_drawn_number = np.mean(tokens)
variance_drawn_number = np.var(tokens)

# Step 4: Create and Visualize Plots
# Plot 1: Histogram
plt.figure(figsize=(12, 6))
plt.subplot(1, 2, 1)
plt.hist(tokens, bins=20, density=True, alpha=0.6, color='b', label='Drawn Numbers (Histogram)')
plt.axvline(mean_drawn_number, color='r', linestyle='dashed', linewidth=2, label='Sample Mean')

# Add shaded region for numbers between 30 and 70
plt.axvspan(30, 70, alpha=0.3, color='gray', label='30 to 70')

plt.xlabel('Drawn Numbers')
plt.ylabel('Probability Density')
plt.legend()
plt.title('Lucky Draw Contest - Histogram')

# Display the plots
plt.tight_layout()
plt.show()

# Print the calculated values
print("Problem Statement:")
print("You are organizing a Lucky Draw Contest with 100 participants. Each participant draws a token numbered from 1 to 100.")
print("\nResults:")
print(f"Probability of drawing a number less than or equal to 10: {probability_less_than_10:.2f}")
print(f"Probability of drawing an even number: {probability_even_number:.2f}")
print(f"Probability of drawing a number between 30 and 70: {probability_between_30_and_70:.2f}")
print(f"Mean of drawn numbers: {mean_drawn_number:.2f}")
print(f"Variance of drawn numbers: {variance_drawn_number:.2f}")
```

Central Limit Theorem

```python
import numpy as np
import matplotlib.pyplot as plt

# Parameters for the population distribution (1 to 100)
population_min = 1
population_max = 100

# Number of samples and sample size
num_samples = 1000
sample_size = 30

# Create an array to store sample means
sample_means = []

# Generate data and calculate sample means
for _ in range(num_samples):
    sample = np.random.randint(population_min, population_max + 1, sample_size)
    sample_mean = np.mean(sample)
    sample_means.append(sample_mean)

# Plot the sample means
plt.figure(figsize=(8, 6))
plt.hist(sample_means, bins=30, density=True, alpha=0.6, color='b', label='Sample Means')
plt.xlabel('Sample Mean')
plt.ylabel('Probability Density')
plt.title('Central Limit Theorem Demonstration with Integer Data (1 to 100)')
plt.legend()
plt.grid(True)

# Fit and plot a normal distribution for comparison
sample_mean_mean = np.mean(sample_means)
sample_mean_std = np.std(sample_means)
x = np.linspace(sample_mean_mean - 3 * sample_mean_std, sample_mean_mean + 3 * sample_mean_std, 1000)
normal_pdf = 1 / (sample_mean_std * np.sqrt(2 * np.pi)) * np.exp(-(x - sample_mean_mean)**2 / (2 * sample_mean_std**2))
plt.plot(x, normal_pdf, color='r', label='Approx. Normal Distribution (CLT)')

plt.legend()
plt.show()
```

Linear Regression

```python
import numpy as np
import matplotlib.pyplot as plt

# Data
x = np.array([1, 2, 3, 4, 5])
y = np.array([2, 4, 5, 4, 5])

# Calculate the linear regression
slope, intercept = np.polyfit(x, y, 1)
regression_line = slope * x + intercept

# Calculate the least squares error
error = np.sum((y - (slope * x + intercept)) ** 2)

# Plot the data points and the best-fit line
plt.scatter(x, y, label='Data Points', color='b')
plt.plot(x, regression_line, label='Linear Regression', color='r')
plt.xlabel('X')
plt.ylabel('Y')
plt.title('Linear Regression and Least Squares Error')
plt.legend()

# Plot dots and lines connecting actual and estimated points
for i in range(len(x)):
    plt.plot([x[i], x[i]], [y[i], slope * x[i] + intercept], 'g--', lw=2)

plt.grid()
plt.show()

print(f"Least Squares Error: {error:.2f}")
```

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm

# Generate sample data (Restricted Range: Scores 50-90)
np.random.seed(42)  # For reproducibility
exam_scores = np.random.normal(loc=70, scale=10, size=100)  # Mean=70, Std=10, N=100
exam_scores = np.clip(exam_scores, 50, 90)  # Restrict values between 50 and 90

# Sort the scores for better display in table
exam_scores_sorted = np.sort(exam_scores)

# Print raw score data in tabular form
print("\nRAW SCORE DATA (Sorted):")
print("="*60)
print("Index | Score | Index | Score | Index | Score | Index | Score | Index | Score | Index | Score")
print("-"*60)
for i in range(0, 100, 6):  # Print 4 columns
    row = []
    for j in range(6):
        if i+j < 100:
            row.append(f"{i+j:5d} | {exam_scores_sorted[i+j]:5.1f}")
    print(" | ".join(row))

# Define bin edges (50-90 with bin width 5)
bin_size = 5
bin_edges = np.arange(50, 90 + bin_size, bin_size)

# Compute histogram
hist_values, _ = np.histogram(exam_scores, bins=bin_edges)

# Compute relative frequency and PDF values
total_samples = len(exam_scores)
relative_frequencies = hist_values / total_samples
pdf_values = relative_frequencies / bin_size

# Compute bin centers
bin_centers = (bin_edges[:-1] + bin_edges[1:]) / 2

# Theoretical Normal PDF for comparison
x = np.linspace(50, 90, 100)
theoretical_pdf = norm.pdf(x, loc=70, scale=10)

# Print formulas
print("\n\nFORMULAS USED:")
print("="*60)
print("1. Relative Frequency = Frequency / Total Samples")
print("2. PDF Value = Relative Frequency / Bin Width")
print("3. Theoretical PDF = (1/((2))) * e^(-(x-)/(2))")
print("   where  = mean,  = standard deviation")
print("4. Sample Mean = (x_i)/n")
print("5. Sample Std = (((x_i - mean))/(n-1))")

# Display Numerical Data
print("\n\nHISTOGRAM TO PDF CONVERSION:")
print("="*90)
header = f"{'Score Range':<12} | {'Frequency':>9} | {'Rel Freq':>9} | {'PDF Value':>9} | {'Theoretical PDF':>15}"
print(header)
print("-"*90)
for i in range(len(hist_values)):
    theor_pdf = norm.pdf(bin_centers[i], loc=70, scale=10)
    print(f"{bin_edges[i]:2d}-{bin_edges[i+1]:2d} | {hist_values[i]:9d} | {relative_frequencies[i]:9.4f} | {pdf_values[i]:9.4f} | {theor_pdf:15.4f}")

# Summary Statistics with formulas
print("\n\nSUMMARY STATISTICS:")
print("="*60)
print(f"Sample size (n): {total_samples}")
print(f"Bin size (h): {bin_size} points")
print(f"Sample mean (): {np.mean(exam_scores):.2f} (Formula:  = (x_i)/n)")
print(f"Sample std (): {np.std(exam_scores):.2f} (Formula:  = (((x_i - ))/(n-1)))")
print(f"Total relative frequencies sum: {np.sum(relative_frequencies):.4f} (Should be 1)")
print(f"Integral of estimated PDF (pdf*h): {np.sum(pdf_values * bin_size):.4f} (Should be 1)")

# Plotting
plt.figure(figsize=(14, 6))

# Histogram Plot
plt.subplot(1, 2, 1)
bars = plt.bar(bin_centers, hist_values, width=bin_size*0.9, color='skyblue', edgecolor='navy', alpha=0.7)
plt.title(f"Exam Score Histogram\n(Bin Size = {bin_size} points)", pad=20)
plt.xlabel("Score Range")
plt.ylabel("Frequency")
plt.grid(axis='y', linestyle='--', alpha=0.7)

# PDF Plot
plt.subplot(1, 2, 2)
plt.bar(bin_centers, pdf_values, width=bin_size*0.9, color='salmon', edgecolor='darkred', alpha=0.7, label="Estimated PDF")
plt.plot(x, theoretical_pdf, 'k-', linewidth=2, label="Theoretical Normal PDF")
plt.title("Probability Density Function Comparison", pad=20)
plt.xlabel("Score")
plt.ylabel("Density")
plt.legend()
plt.grid(axis='y', linestyle='--', alpha=0.7)

plt.tight_layout()
plt.show()
```
