```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
from imblearn.over_sampling import SMOTE, RandomOverSampler
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score, confusion_matrix, classification_report
```

```python
df = pd.read_csv("/content/Breast_cancer_dataset.csv")
```

```python
df.head(5)
```

```python
df.info()
```

```python
df.drop(["id", "Unnamed: 32"], axis=1, inplace=True)
```

```python
df
```

```python
df['diagnosis'].value_counts()
```

```python
df['diagnosis'] = df['diagnosis'].map({'M':1, 'B':0})
```

```python
df
```

```python
X = df.drop('diagnosis', axis=1)
Y = df['diagnosis']
```

```python
print(X.shape, Y.shape)
```

```python
smote = SMOTE(random_state=42)
X_resampled, Y_resampled = smote.fit_resample(X, Y)
```

```python
print(X_resampled.shape, Y_resampled.shape)
```

```python
X_train, X_test, Y_train, Y_test = train_test_split(X_resampled, Y_resampled, test_size=0.2, random_state=42)
```

```python
# prompt: rather than showing all the models separately how can i show it on table

import pandas as pd
# ... (Your existing code)

models = {
    "Logistic Regression": LogisticRegression(),
    "Decision Tree": DecisionTreeClassifier(),
    "Random Forest": RandomForestClassifier()
}

results = []
for name, model in models.items():
    model.fit(X_train, Y_train)
    yhat_predict_train = model.predict(X_train)
    yhat_predict_test = model.predict(X_test)

    train_accuracy = accuracy_score(Y_train, yhat_predict_train)
    test_accuracy = accuracy_score(Y_test, yhat_predict_test)

    results.append([name, train_accuracy, test_accuracy])

results_df = pd.DataFrame(results, columns=["Model", "Training Accuracy", "Test Accuracy"])
results_df
```
