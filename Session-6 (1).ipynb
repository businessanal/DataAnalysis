### Bar plot

## Understanding Matplotlib workflow steps:


```python
# 0. import matplotlib and get it ready for plotting in jupyter
import matplotlib.pyplot as plt
import pandas as pd
%matplotlib inline

# 1. Prepare data
x = [1,2,3,4]
y = [4,5,3,2]

# 2. setup the plot
fig, ax = plt.subplots(figsize=(8,6))

# 3. Plot the data
ax.plot(x,y,label='legend_name',
        color='green',
        linewidth=4,
        alpha=0.5, 
        marker="*", 
        markersize=15, 
        markerfacecolor='red',
       markeredgewidth=2,
       markeredgecolor='blue')

# 4. customize the plot
ax.set(title='Simple plot',
       xlabel='x-axis',
       ylabel='y-axis')

# 5. save the figure
fig.savefig('sample-plot.png',
            dpi=200)
ax.legend();
```

## Seaborn

```python
import seaborn as sns

# Seaborn builtin datasets
tips = sns.load_dataset('tips')
# first 5 rows of datasets
tips
```

```python
# changing the aaggregation statistic to compute each bin
sns.histplot(data=tips,
             x='sex', 
             stat="percent", 
             );
```

```python
# changing the aaggregation statistic to compute each bin
sns.histplot(data=tips,
             x='sex',
             stat='density');
plt.show()
```

```python
sns.histplot(data=tips,
             y='tip',
             discrete=True);
```

```python
sns.scatterplot(data=tips,
                x="total_bill",
                y="tip");
```

```python
sns.scatterplot(data=tips,
                x="total_bill",
                y="tip",
                hue="sex");
#hue parameter take column name for colour encoding
```

```python
sns.scatterplot(
    data=tips, x="total_bill",
    y="tip",
    hue="size",
    size="size",
    sizes=(20, 200),
    legend="full"
);
```

```python
flights = sns.load_dataset('flights')
flights
```

```python
sns.lineplot(data=flights, 
             x="year",
             y="passengers", 
             hue="month",
             style='month', 
             markers=True);#hue=color encoding, style:'-.-.-.-', marker:*,.
```

```python
sns.lineplot(data=flights, 
             x="year",
             y="passengers", 
             hue="month",
             style='month',
             markers=True);
plt.show ()
```

```python
sns.barplot(x='sex',
            y='total_bill',
            data=tips);# x-category , y=numeric
```

```python
penguins = sns.load_dataset("penguins")
sns.pairplot(data=penguins);
```
